```python
# 基底関数：多項式
k = 0
x = np.linspace(-1, 1, 100)
x_prime = -0.5
for i in range(1, 11):
    y = x**i
    k += x**i * x_prime**i
    plt.plot(x, y)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()
plt.plot(x, k)
plt.scatter(x_prime, -0.4, marker='x', clip_on=False)
plt.xlim(-1, 1)
plt.ylim(-0.4, 1)
```

```python
# 基底関数：ガウス基底関数
k=0
x = np.linspace(-1, 1, 100)
x_prime = 0
for mu in np.linspace(-1, 1, 11):
    y = exp(-(x-mu)**2/ (2*(0.2)**2))
    k += exp(-(x-mu)**2 / (2*(0.2)**2)) * exp(-(x_prime-mu)**2 / (2*(0.2)**2))
    plt.plot(x, y)
plt.xlim(-1, 1)
plt.ylim(0, 1)
plt.show()
plt.plot(x, k)
plt.scatter(x_prime, 0, marker='x', clip_on=False)
plt.xlim(-1, 1)
plt.ylim(0, 2)
```

```python
# 基底関数：シグモイド関数
k=0
x = np.linspace(-1, 1, 100)
x_prime = 0
for mu in np.linspace(-1, 1, 11):
    y = 1 / (1+exp(-((x-mu) / 0.2)))
    k += 1 / (1+exp(-((x-mu) / 0.2))) * 1 / (1+exp(-((x_prime-mu) / 0.2)))
    plt.plot(x, y)
plt.xlim(-1, 1)
plt.ylim(0, 1)
plt.show()
plt.plot(x, k)
plt.scatter(x_prime, 0, marker='x', clip_on=False)
plt.xlim(-1, 1)
plt.ylim(0, 6)
```

6.3 RBFネットワーク

```python
N_train = 10
N_test = 100
x_train = np.linspace(0, 1, N_train) + np.random.normal(0, 0.05, N_train)
x_test = np.linspace(0, 1, N_test)
t_train = np.sin(2*pi*x_train) + np.random.normal(0, 0.1, N_train)
t_test = np.sin(2*pi*x_test)
```

```python
k = []
std = 0.1
gauss = lambda x, mu, std: np.exp(-(x-mu)**2 / (2*(std)**2))
for i ,mu in zip(range(N_train), x_train):
    k.append(gauss(x_test, mu, std))
k /= np.sum(k, axis=0, keepdims=True)
k = np.array(k)
y = np.sum(k*t_train.reshape(-1, 1), axis=0)

plt.scatter(x_train, t_train)
plt.plot(x_test, t_test)
plt.plot(x_test, y)
```
